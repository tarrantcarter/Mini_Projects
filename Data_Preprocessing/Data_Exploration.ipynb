```
# Set up environment
import math
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# supress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")
```

```
data_url = 'https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/Doctor.csv'

visits = pd.read_csv(data_url)

visits.head()
```

```
visits.describe()
```

```
sns.distplot(visits['doctor'])
plt.show()

print(stats.describe(visits['doctor']))
print("doctor: ", stats.shapiro(visits['doctor']))
```

```
sns.distplot(visits['children'])
plt.show()

print(stats.describe(visits['children']))
print("children: ", stats.shapiro(visits['children']))
```

```
sns.distplot(visits['access'])
plt.show()

print(stats.describe(visits['access']))
print("access: ", stats.shapiro(visits['access']))
```

```
sns.distplot(visits['health'])
plt.show()

print(stats.describe(visits['health']))
print("health: ", stats.shapiro(visits['health']))
```
