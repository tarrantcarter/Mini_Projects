```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
```

#Normal Distribution

```
results = []
for num_trials in range(1,1000):
    trials = np.random.normal(0,1, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})

df.plot()
```

```
means = [mean(np.random.normal(0,1, 1000)) for _ in range(1000)]


plt.hist(means, bins=30)
plt.show()
```

#Binomial

```
results = []
for num_trials in range(1,1000):
    trials = np.random.binomial(100,.5,num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})

df.plot()
```

```
means = [mean(np.random.binomial(100,.5, 1000)) for _ in range(1000)]

plt.hist(means, bins=30)
plt.show()
```
