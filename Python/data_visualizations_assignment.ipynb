```
import pandas as pd
houseprices = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')

houseprices.shape

houseprices.head(30)
```

# Illustrate the distribution of saleprice.

```
import seaborn as sns 
sns.set()

# suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")

saleprice = sns.distplot(houseprices['saleprice'])
```

# Illustrate the distribution of the lotarea.

```
lotarea = sns.distplot(houseprices['lotarea'])
```

# Use a box plot to illustrate how overallcond affects the sale price.

```
sns.boxplot(x='overallcond', y='saleprice', data=houseprices)
```

# On the same box plot, highlight the difference between the sale price for houses that have central air conditioning and those that do not.

```
sns.boxplot(x='overallcond', y='saleprice', hue='centralair', data=houseprices)
```

# Is there a correlation between lotarea and the saleprice?

```
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(40,40))
sns.heatmap(houseprices.corr(), annot=True, linewidths=1, ax=ax)
```
