```
open_porch = houseprices.iloc[(houseprices['openporchsf'] >= 1).values]
no_open_porch = houseprices.iloc[(houseprices['openporchsf'] < 1).values]

open_porch.info()
no_open_porch.info()
```

```
from scipy import stats

stats.ttest_ind(open_porch['saleprice'], no_open_porch['saleprice'])
```
