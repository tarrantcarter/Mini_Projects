# Find all successful documentary projects and sort them by the amount pledged. Print the top 10 highest pledges.

```
successful_docs = (ks_db['category'] == 'Documentary') & (ks_db['state'] == 'successful')
successful_docs = ks_db.iloc[successful_docs.values]
most_pledged_docs = successful_docs.sort_values(by='pledged', ascending=False)
most_pledged_docs.head(10)
```

# Create a new column named average_per_backer and set its value to the total amount pledged divided by the number of backers. What happened to the rows with zero backers? How can this be dealt with?

```
ks_db_new_column = ks_db.assign(average_per_backer = ks_db.usd_pledged / ks_db.backers)
ks_db_new_column.head(20)
```

The rows with 0 backers have NaN values in the 'average_per_backer' column. We can deal with this by assigning 0 for the NaN values in the 'average_per_backer' column. Or we can remove all rows with 0 backers and then create our 'average_per_backer' column.

# Drop all rows with zero backers, then repeat the previous exercise.

```
at_least_one_backer = ks_db['backers'] >= 1
ks_db_updated = ks_db.iloc[at_least_one_backer.values]
ks_db_updated = ks_db_updated.assign(average_per_backer = ks_db_updated.usd_pledged / ks_db_updated.backers)
ks_db_updated.head(20)
```

# Merge these two DataFrames into a single DataFrame: nycflights13.

```
flights = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/flights.csv')
airlines = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/airlines.csv')

nycflights13 = pd.merge(flights, airlines)
nycflights13.info()
nycflights13.head(30)
```

# **census**

# Merge these two DataFrames into a single DataFrame called census.

```
population = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')
regions = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')

census = pd.merge(population, regions)
census.info()
census.head(51)
```

# Reshape census so that one column contains all population measures, and another contains all of the year attributes.

```
census_melted = pd.melt(frame=census, id_vars=['state', 'region', 'division'], value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016'], var_name='year', value_name='population')
census_melted
```

# Group the data by year and summarize it.

```
census_melted.groupby(["year"]).describe()
```

# Group the data by region, division, and year. Summarize it.

```
pd.set_option('display.max_rows', None) 

census_grouped = census_melted.groupby(['region', 'division', 'year']).describe()
census_grouped
```
